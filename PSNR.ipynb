```python
import skimage
import os
path_pred='./Results/Vid4/foliage_4x/'
path_gt='./Vid4/foliage/'
```

```python
pred = os.listdir(path_pred)
```

```python
gt = os.listdir(path_gt)
```

```python
pred = sorted(pred)
gt = sorted(gt)
```

```python
import re

def tryint(s):
    try:
        return int(s)
    except:
        return s

def alphanum_key(s):
    """ Turn a string into a list of string and number chunks.
        "z23a" -> ["z", 23, "a"]
    """
    return [ tryint(c) for c in re.split('([0-9]+)', s) ]

def sort_nicely(l):
    """ Sort the given list in the way that humans expect.
    """
    l.sort(key=alphanum_key)
```

```python
sort_nicely(pred)
```

```python
average_psnr = 0.0
average_ssim = 0.0
count = 1
for i in range(0,len(gt)):
    predicted = skimage.io.imread(path_pred+pred[i])
    golden = skimage.io.imread(path_gt+gt[i])
    psnr = skimage.measure.compare_psnr(golden,predicted)
    ssim = skimage.measure.compare_ssim(golden,predicted,multichannel=True)
    print('#####################')
    print('PSRN:' + str(psnr))
    print('SSIM:' + str(ssim))
    average_ssim += ssim
    average_psnr +=psnr
    count+=1
    
print('######################################')

print('Average PSNR',average_psnr/count)
print('Average SSIM',average_ssim/count)
```
